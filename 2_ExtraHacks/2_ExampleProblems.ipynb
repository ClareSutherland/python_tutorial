```python
# Now let's duplicate it 50 times
import shutil

samplesize = 50
for subjnumber in range(samplesize):
    outfilename = "myexp/generated/subject0" + str(subjnumber)
    shutil.copyfile("myexp/experiment_final.xml", outfilename, follow_symlinks=True)  
```

## Aggregating CSV File with Pandas

```python
import glob
import pandas as pd

allFiles = glob.glob("./data/*.csv")

dataframe = pd.DataFrame()
mini_list = []
for file in allFiles:
    this_participant = pd.read_csv(file,index_col=None, header=0)
    mini_list.append(this_participant)
dataframe = pd.concat(mini_list)

dataframe
```

```python
print(dataframe['ReactionTime'].mean())

dataframe.groupby('ID').mean()
```

```python
dataframe2 = dataframe.drop(['Age', 'ID'], axis = 1)
dataframe2.groupby('Stimulus').mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

dataframe2.groupby('Stimulus').mean().plot()
```
